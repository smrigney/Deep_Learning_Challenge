```python
# Import our dependencies
from sklearn.model_selection import train_test_split as TTS
from sklearn.preprocessing import StandardScaler as SS
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
df = pd.read_csv("charity_data.csv")
df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
df = df.drop(['EIN', 'NAME'], axis =1)

df.head()
```

```python
# Determine the number of unique values in each column.
df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_type_cnt = df.value_counts(df['APPLICATION_TYPE'])
app_type_cnt
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(app_type_cnt[app_type_cnt<1000].index)

# Replace in dataframe
for app in application_types_to_replace:
    df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_type_cnt = df.value_counts(df['CLASSIFICATION'])
class_type_cnt
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_type_cnt = class_type_cnt[class_type_cnt > 1]
class_type_cnt 
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(class_type_cnt[class_type_cnt<1200].index)

# Replace in dataframe
for cls in classifications_to_replace:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df_dummy = pd.get_dummies(df)
df_dummy
```

```python
# Split our preprocessed data into our features and target arrays
df_target = df_dummy['IS_SUCCESSFUL'].values

df_feat = df_dummy.drop(['IS_SUCCESSFUL'], axis = 1).values
```

```python
# Split the preprocessed data into a training and testing dataset

xtrain, xtest, ytrain, ytest = TTS(df_feat, df_target, random_state = 78)
```

```python
# Create a StandardScaler instances
scaler = SS()

# Fit the StandardScaler
X_scaler = scaler.fit(xtrain)

# Scale the data
xtrain_scaled = X_scaler.transform(xtrain)
xtest_scaled = X_scaler.transform(xtest)
```

```python
# Create a method that creates a new Sequential model with hyperparameter options
num_features = len(xtrain)
hidden_nodes_1 = 8
hidden_nodes_2 = 5
hidden_nodes_3 = 5

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense
       (units=hidden_nodes_1,
        input_shape = (66,),
        input_dim= num_features, 
        activation = "relu"
       ))

# Second hidden layer
nn.add(tf.keras.layers.Dense
      (units = hidden_nodes_2, 
       activation = "relu"
      ))

# Third hidden layer
nn.add(tf.keras.layers.Dense
      (units = hidden_nodes_3, 
       activation = "relu"
      ))

# Output layer
nn.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

#Check the structure
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
```

```python
print(xtrain_scaled.shape)
print(ytrain.shape)
```

```python
# Train the model
fit_model = nn.fit(xtrain_scaled, ytrain, epochs = 100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(xtest_scaled,ytest,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity_Optimization.h5")
```
